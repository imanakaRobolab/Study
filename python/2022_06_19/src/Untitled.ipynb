```python
import pandas
import pandas_datareader
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

```python
df1 = web.DataReader('GOOG', 'yahoo', start='2021-06-17', end='2022-06-17')
df2 = web.DataReader('AAPL', 'yahoo', start='2021-06-17', end='2022-06-17')
df3 = web.DataReader('FB-USD', 'yahoo', start='2021-06-17', end='2022-06-17')

plt.plot(df1.index,df1['Open'])
plt.plot(df2.index,df2['Open'])
plt.plot(df3.index,df3['Open'])
plt.legend(['GOOG','AAPL','FB-USD'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Open of stack')

plt.show()
```
